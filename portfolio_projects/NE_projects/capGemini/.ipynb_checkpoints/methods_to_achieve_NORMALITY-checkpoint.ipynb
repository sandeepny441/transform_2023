# Example of data that is NOT-NORMALLY distributed

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set random seed for reproducibility
np.random.seed(0)

# Generate original_data with exponential distribution (right-skewed)
original_data = np.random.exponential(scale=100, size=1000)

# Plot original_data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Data')
plt.hist(original_data, bins=30, edgecolor='black')
plt.xlabel('Income')
plt.ylabel('Frequency')

# QQ-Plot for original data
plt.subplot(1, 2, 2)
stats.probplot(original_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of Original Data')
plt.show()
```

# Log Transformation

```python
import numpy as np

log_data = np.log1p(original_data)  # log1p computes log(x + 1)
# QQ-Plot for original data
plt.subplot(1, 2, 2)
stats.probplot(log_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of log_data')
plt.show()
```

# Square Root Transformation

```python
sqrt_data = np.sqrt(original_data)
# QQ-Plot for TRANSFORMED data
plt.subplot(1, 2, 2)
stats.probplot(sqrt_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of TRANSFORMED data')
plt.show()
```

# Box-Cox Transformation

```python
from scipy.stats import boxcox

boxcox_data, lambda_value = boxcox(original_data)
# QQ-Plot for TRANSFORMED data
plt.subplot(1, 2, 2)
stats.probplot(boxcox_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of TRANSFORMED data')
plt.show()
```

# Yeo-Johnson Transformation

```python
from scipy.stats import yeojohnson

yeojohnson_data, lambda_value = yeojohnson(original_data)
# QQ-Plot for TRANSFORMED data
plt.subplot(1, 2, 2)
stats.probplot(yeojohnson_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of TRANSFORMED data')
plt.show()
```

# Z score Transformation

```python
from scipy.stats import zscore

z_data = zscore(original_data)

# QQ-Plot for TRANSFORMED data
plt.subplot(1, 2, 2)
stats.probplot(z_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of TRANSFORMED data')
plt.show()

```

# Quantile Transformation

```python
from sklearn.preprocessing import QuantileTransformer

transformer = QuantileTransformer(output_distribution='normal')
transformed_data = transformer.fit_transform(original_data.reshape(-1, 1))
# QQ-Plot for TRANSFORMED data
plt.subplot(1, 2, 2)
# Using numpy's flatten method
transformed_data_flat = transformed_data.flatten()

# # OR using numpy's ravel method
# transformed_data_flat = transformed_data.ravel()

stats.probplot(transformed_data_flat, dist="norm", plot=plt)
plt.title('Q-Q Plot of TRANSFORMED data')
plt.show()
```

# Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(original_data.reshape(-1, 1))

# QQ-Plot for TRANSFORMED data
plt.subplot(1, 2, 2)
# Using numpy's flatten method
transformed_data_flat = scaled_data.flatten()
stats.probplot(transformed_data_flat, dist="norm", plot=plt)
plt.title('Q-Q Plot of TRANSFORMED data')
plt.show()
```
