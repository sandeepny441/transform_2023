```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, kstest, anderson, normaltest, skew, kurtosis

# Generate normally distributed age data
age = np.random.normal(loc=40, scale=10, size=1000)

# Generate normally distributed savings_amount data
savings_amount = np.random.normal(loc=50000, scale=10000, size=1000)

# Create a DataFrame
df = pd.DataFrame({
    'Age': age,
    'SavingsAmount': savings_amount
})

df
```

# Histogram

```python
import matplotlib.pyplot as plt

# Generate your data
# data = [...]

data = df['Age']

# Create histogram
plt.hist(data, bins=30, density=True)
plt.title('Histogram')
plt.show()
```

# QQ plot

```python
import scipy.stats as stats

# Generate your data
# data = [...]

# Create Q-Q plot
stats.probplot(data, dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

# Box plot 

```python
# Generate your data
# data = [...]

# Create boxplot
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

# KDE plot

```python
import seaborn as sns

# Generate your data
# data = [...]

# Create KDE plot
sns.kdeplot(data)
plt.title('KDE Plot')
plt.show()
```
