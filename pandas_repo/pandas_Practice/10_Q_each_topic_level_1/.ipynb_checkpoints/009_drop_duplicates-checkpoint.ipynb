```python
import pandas as pd
data = {
'Product_ID': [1, 2, 3, 4, 5, 6, 1, 2, 3, 4],
'Product_Name': ['Toothpaste', 'Shampoo', 'Laptop', 'TV', 'Coffee Maker', 'Blender', 'Toothpaste', 'Shampoo', 'Laptop', 'TV'],
'Category': ['Personal Care', 'Personal Care', 'Electronics', 'Electronics', 'Kitchen Appliances', 'Kitchen Appliances', 'Personal Care', 'Personal Care', 'Electronics', 'Electronics'],
'Price': [5.0, 10.0, 1000.0, 600.0, 50.0, 30.0, 5.0, 10.0, 1000.0, 600.0],
'Units_Sold': [100, 50, 20, 15, 30, 25, 110, 55, 22, 17]
}

target_df = pd.DataFrame(data)

target_df
```

```python
# How would you use the drop_duplicates function to 
# remove all rows that have identical values across all columns?\


target_df.drop_duplicates()
```

```python
# If your DataFrame contains columns 'Product_ID', 
# 'Product_Name', and 'Price', 
# how would you use drop_duplicates to keep only the unique 'Product_ID' rows, while retaining the first occurrence of each 'Product_ID'?
target_df.drop_duplicates(['Product_ID'], keep = 'first')
```

```python
# Is it possible to use drop_duplicates to 
# remove duplicates based on multiple columns, 
# say 'Product_ID' and 'Price'? If so, how would you do it?

target_df.drop_duplicates(subset = ['Product_ID', 'Price'])
```

```python
# How would you use drop_duplicates to keep the 
# last occurrence of each duplicate row based on 
# 'Product_ID' and remove the earlier ones?
# Can you specify

target_df.drop_duplicates(['Product_ID'], keep = False)
```

```python

# Subset Prioritization: Imagine you have three columns 
# 'Product_ID', 'Product_Name', and 'Price'. If you only want to remove duplicates
# based on 'Product_ID' and 'Product_Name' while ignoring 'Price', how would you do it using drop_duplicates?

target_df.drop_duplicates(['Product_ID', 'Product_Name'], keep = "first")
```

```python
# Unique Rows Count: How would you count the number of unique 
# rows in your DataFrame after considering duplicates 
# based on the columns 'Product_ID' and 'Price'?

target_df.drop_duplicates(['Product_ID', 'Product_Name'], keep = "first").shape[0]
```

```python
# Inplace Alteration: By default, 
# the drop_duplicates function returns a new DataFrame. 
# How would you modify the original DataFrame in place to remove duplicates?

target_df.drop_duplicates(inplace = True)
```

```python
# Case Sensitive Check: 
# Is the drop_duplicates function case sensitive by default 
# when it checks for duplicates in string columns? How would you confirm this?

# target_df.iloc[9, 1] = 'shampoo'
# target_df
target_df.drop_duplicates(['Product_Name'])

# Answer is CASE Sensitrive
```

```python
# Resetting Index: After dropping duplicates, 
# the DataFrame's index may no longer be continuous. 
# How would you reset the index after using drop_duplicates?

target_df.drop_duplicates().reset_index()
```
