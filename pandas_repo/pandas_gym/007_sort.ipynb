```python
import pandas as pd
import numpy as np

# create a sample dataframe
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 4],
    'B': [np.nan, 2, 4, 6, 5],
    'C': [2399, 12, 3, 4, 4]
})

df
```

```python
df.sort_values(by = 'C')
```

```python
df
```

```python
df.sort_values(by = 'C')
```

```python
df.sort_values(by = 'C', inplace = True)
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df
```

```python
df.reindex()
```

```python
df.sort_values(by = 'B')
```

```python
df.sort_values(by = 'B').reindex()
```

```python
# sort by column 'A'
df.sort_values(by='A')  # in ascending order
```

```python
# sort by column 'A' in descending order
df.sort_values(by='A', ascending=False)
```

```python
# sort by column 'A' in descending order
df.sort_values(by='A', ascending=False, na_position = 'first')
```

```python
# sort by column 'A' and 'B'
df.sort_values(by=['A', 'B'])
```
