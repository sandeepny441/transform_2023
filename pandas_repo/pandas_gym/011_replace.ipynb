```python
import pandas as pd

data = {
    'Names': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'David Lee'],
    'Address': ['1234 Elm St', '5678 Oak St', '9101 Pine St', '1121 Maple St']
}

df = pd.DataFrame(data)
df
```

```python
# df.replace({'column_name': r'regex_pattern'}, {'column_name': new_value}, regex=True)
```

```python
df = df.replace({'Names': r'Smith$'}, {'Names': 'Jones'}, regex=True)
df = df.replace({'Address': r'Elm|Oak'}, {'Address': 'Birch'}, regex=True)
df
```
