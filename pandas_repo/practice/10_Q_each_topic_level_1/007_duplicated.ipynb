```python
import pandas as pd
data = {
'Product_ID': [1, 2, 3, 4, 5, 6, 1, 2, 3, 4],
'Product_Name': ['Toothpaste', 'Shampoo', 'Laptop', 'TV', 'Coffee Maker', 'Blender', 'Toothpaste', 'Shampoo', 'Laptop', 'TV'],
'Category': ['Personal Care', 'Personal Care', 'Electronics', 'Electronics', 'Kitchen Appliances', 'Kitchen Appliances', 'Personal Care', 'Personal Care', 'Electronics', 'Electronics'],
'Price': [5.0, 10.0, 1000.0, 600.0, 50.0, 30.0, 5.0, 10.0, 1000.0, 600.0],
'Units_Sold': [100, 50, 20, 15, 30, 25, 110, 55, 22, 17]
}

target_df = pd.DataFrame(data)
target_df
```

```python
# How would you identify if there are any duplicate rows 
# in the DataFrame? What function would you use?


target_df.duplicated().any()
```

```python
# If the DataFrame has a column for 'Product_ID', 
# how would you find out if there are any duplicate 'Product_ID' entries?

target_df['Product_ID'].duplicated().any()
```

```python
# How would you remove duplicate rows in the DataFrame? 
# Which function can accomplish this?
target_df.drop_duplicates()
```

```python
target_df.drop_duplicates(subset = 'Product_Name')
```

```python
# If the DataFrame has columns 'Product_ID' and 'Product_Category', how would you remove rows where only 'Product_ID' is duplicated but keep the first occurrence?
target_df['Product_ID'].drop_duplicates()
```

```python
target_df.drop_duplicates(subset = 'Product_ID')
```

```python
target_df.drop_duplicates(subset = 'Product_ID', keep = "first")
```

```python
target_df.drop_duplicates(subset = 'Product_ID', keep = False)
```

```python
target_df.drop_duplicates(subset = 'Product_ID', keep = 'last')
```

```python
#  How can you get a DataFrame with only the rows that are duplicated, 
# excluding the first occurrence of each duplicate?

target_df[target_df.duplicated()]
```

```python
# How would you count the number of duplicate rows in the DataFrame?
target_df.duplicated().count()
```

```python
target_df.duplicated().sum()
```

```python
target_df['Product_ID'].count()
```

```python
target_df['Product_ID'].count()
```

```python
# If you wanted to flag rows that are duplicates 
# with a new column called 'Is_Duplicate', how would you achieve this?
target_df['Is_Duplicate'] = False  # Initializes the column with False
target_df.loc[target_df['Product_ID'].duplicated(), 'Is_Duplicate'] = True
```
