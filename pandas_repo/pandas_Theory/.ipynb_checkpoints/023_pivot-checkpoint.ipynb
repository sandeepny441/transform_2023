```python
import pandas as pd

data = {
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 50, 300, 10]
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot(index='Month', columns='Product', values='Sales')
pivot_df
```

```python
import pandas as pd

data = {
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 50, 300, 10]
}

df = pd.DataFrame(data)
df

```

```python
# Using pivot_table
pivot_df = df.pivot(index='Month', columns='Product', values='Sales')
pivot_df
```

```python
# Using pivot_table
pivot_df = df.pivot_table(index='Month', columns='Product', values='Sales', aggfunc='sum')
pivot_df
```

```python
# aggregations
```

```python
data = {
    'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Mar'],
    'Product': ['A', 'A', 'B', 'A', 'B', 'A', 'A', 'B'],
    'Sales': [100, 150, 200, 50, 300, 10, 20, 40]
}

df = pd.DataFrame(data)

# Using pivot_table with aggregation
pivot_df = df.pivot_table(index='Month', columns='Product', values='Sales', aggfunc='sum')
print(pivot_df)
```

```python
# mutliple aggregations
```

```python
pivot_df = df.pivot_table(index='Month', columns='Product', values='Sales', aggfunc=['sum', 'mean'])
print(pivot_df)
```
